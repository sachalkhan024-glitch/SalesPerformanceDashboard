```python
#setting right datatype
orders['order_purchase_timestamp']=pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at']=pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_customer_date']=pd.to_datetime(orders['order_delivered_customer_date'])
```

```python
#merging data sets into one dataframe
df=orders.merge(order_items,on='order_id',how='inner')
df=df.merge(products[['product_id','product_category_name']],on='product_id',how='left')
df=df.merge(customers,on='customer_id', how='left')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#fixing null values
df['order_approved_at']=df['order_approved_at'].fillna(df['order_purchase_timestamp'])
df['product_category_name']=df['product_category_name'].fillna('Unknown')
```

```python
#creating a column of revenue
df['revenue']=df['price']+df['freight_value']
```

```python
#seperating date time in terms of months, years, days, and weekday
df['order_year']=df['order_purchase_timestamp'].dt.year
df['order_month']=df['order_purchase_timestamp'].dt.month
df['order_day']=df['order_purchase_timestamp'].dt.day
df['order_weekday']=df['order_purchase_timestamp'].dt.dayofweek
```

```python
df_2017=df[df['order_purchase_timestamp'].dt.year == 2017].copy()
```

```python
df_2017.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,5))
#monthly sales
monthly_revenue=df_2017.groupby('order_month')['revenue'].sum()
monthly_revenue.plot(kind='line',marker='o',title='Monthy Revenue 2017')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

```python
#top categories
top_categories=df_2017.groupby('product_category_name')['revenue'].sum().sort_values(ascending=False).head(10)
top_categories.plot(kind='bar', title='Top 10 Product Categories by Revenue')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
weekday_revenue=df_2017.groupby('customer_state')['order_id'].nunique().sort_values(ascending=False).head(10)
weekday_revenue.plot(kind='bar', title='Revenue by Day of Week')
plt.xlabel('Weekday (0=Mon, 6=Sun)')
plt.ylabel('Revenue ($)')
plt.show()
```

```python
state_orders = df_2017.groupby('customer_state')['order_id'].nunique().sort_values(ascending=False).head(10)
state_orders.plot(kind='bar', title='Top 10 States by Orders')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# % of revenue from top category
top_cat_rev = df_2017[df_2017['product_category_name'] == 'cama_mesa_banho']['revenue'].sum()
total_rev = df_2017['revenue'].sum()
print(f"Cama Mesa Banho: {top_cat_rev/total_rev*100:.1f}% of total revenue")

# % of orders from SP
sp_orders = df_2017[df_2017['customer_state'] == 'SP']['order_id'].nunique()
total_orders = df_2017['order_id'].nunique()
print(f"SP: {sp_orders/total_orders*100:.1f}% of total orders")
```

# Insights & Recommendations

## Top 3 Insights
1. **November generated the highest revenue** — driven by Black Friday. The business should allocate extra marketing budget to Oct-Nov campaigns.
2. **São Paulo accounts for 39.3% of all orders** — while SP is the largest market, this concentration is a risk. Expand marketing efforts in MG, RJ, and RS to diversify.
3. **Cama Mesa Banho is the top category** but only represents 8.3% of revenue — revenue is well-diversified across categories, so no single category is critical.

## Recommendations
1. **Invest in Black Friday Marketing:** November is clearly the highest-revenue period. Increase ad spend and inventory preparation for Oct-Nov campaigns.
2. **Expand Regional Marketing:** Launch targeted campaigns in MG, RJ, and RS to reduce dependency on SP. A 10% increase in these states would boost total orders by approximately 4-5%.
3. **Promote Top Categories:** Cama Mesa Banho and Beleza Saude are strong performers. Consider cross-selling or bundling these with lower-performing categories.
4. **Monitor December Dip:** December revenue drops despite peak season. Investigate whether this is due to delivery delays, reduced marketing, or customer behavior.
